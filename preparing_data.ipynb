```python
from copy import deepcopy
from sklearn.impute import SimpleImputer
new_df=deepcopy(df)
print(df.iloc[27,38])

new_df.iloc[27,38]=np.nan
print(new_df.iloc[27,38])
print(df.iloc[27,38])
imputer = SimpleImputer(strategy='most_frequent') ####imputing pokoj
data_imputed_array = imputer.fit_transform(new_df)
df_imputed=pd.DataFrame(data=data_imputed_array,columns=new_df.columns)
print(df_imputed.iloc[27,38])
df=df_imputed
df_imputed.to_csv('./res/df_imputed.csv',index=False)
```

```python
##### converting WF to numbers
def stoi(string:str)->int:
    new_str=''
    for character in string:
        if character.isnumeric():
            new_str+=character
        else: 
            break
    return int(new_str)
df['Wf']=df['Wf'].apply(stoi)
print(df['Wf'])
```
