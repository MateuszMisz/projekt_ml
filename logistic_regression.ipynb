```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('./res/df_inne_choosen')
df.drop('Napoj__soki',axis=1,inplace=True)
X=df.drop('Moda',axis=1)
y=df['Moda']
scaler=MinMaxScaler()
X_scaled=scaler.fit_transform(X)
X=pd.DataFrame(X_scaled,columns=X.columns)
```

```python
param_dict={
    'solver':['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'],     
     'C':[0.001, 0.01, 0.1, 1, 10, 100],
    'max_iter':[10,50,100],
    'penalty':['l1', 'l2',None],
    'tol':[0.001, 0.01, 0.1, 1, 10, 100],
    'class_weight':['balanced',None],
    'verbose':[False],
}
```

```python
searcher=GridSearchCV(LogisticRegression(verbose=False),param_grid=param_dict,cv=5,n_jobs=-1,scoring='f1_weighted',verbose=False)
searcher.fit(X,y)
print('najlepsze parametry: ',searcher.best_params_)
print('najlepszy wynik: ',searcher.best_score_)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

lr=LogisticRegression(**{ 'class_weight': 'balanced', 'max_iter': 60, 'penalty': None, 'solver': 'sag', 'tol': 0.01, 'verbose': False})
lr.fit(X,y)
y_pred = lr.predict(X_test)
coef_class_0={}
for i in range(len(X.columns)):
    coef_class_0[X.columns[i]]=lr.coef_[0][i]
print('class0')
for key,value in coef_class_0.items():
    if value != 0: print(f'{key}: {value}')

coef_class_1={}
for i in range(len(X.columns)):
    coef_class_1[X.columns[i]]=lr.coef_[1][i]
print('class1')
for key,value in coef_class_1.items():
    if value != 0: print(f'{key}: {value}')

coef_class_2={}
for i in range(len(X.columns)):
    coef_class_2[X.columns[i]]=lr.coef_[2][i]
print('class2')
for key,value in coef_class_2.items():
    if value != 0: print(f'{key}: {value}')
print(f1_score(y_test,y_pred,average='weighted'))
```

```python
from sklearn.model_selection import cross_val_score
cv_lr=LogisticRegression(**{'C': 10, 'class_weight': None, 'max_iter': 100, 'penalty': 'l1', 'solver': 'liblinear', 'tol': 1, 'verbose': False})
scores=cross_val_score(cv_lr,X,y,cv=5,scoring='f1_weighted',n_jobs=-1,verbose=False)
print(f"Wyniki cross-validation: {scores}")
print(f"rednia f1: {scores.mean()}")
```
